**Задание 1**. Даны 2 строки: *long_phrase* и *short_phrase*. 
Напишите код, который проверяет действительно ли длинная фраза *long_phrase* длиннее короткой *short_phrase*. И выводит *True* или *False* в зависимости от результата сравнения.

```python
long_phrase = 'Насколько проще было бы писать программы, если бы не заказчики'
short_phrase = '640Кб должно хватить для любых задач. Билл Гейтс (по легенде)'
```

```python
long = len(long_phrase)
short = len(short_phrase)
if long > short:
    print(True)
else:
    print(False)
```

**Задание 3**. Разработать приложение для определения знака зодиака по дате рождения. Пример:

Введите месяц: март

Введите число: 6

Вывод:
Рыбы

```python
month = str(input('Введите месяц: '))
day = int(input('Введите число: '))
if (month == 'март' and (20 < day <= 31)) or (month == 'апрель' and (1 <= day <= 19)):
    print('Знак зодиака: Овен\nСтихия: Огонь\nПланета: Марс')
elif (month == 'апрель' and (20 <= day <= 30)) or (month == 'май' and (1 <= day <= 20)):
    print('Знак зодиака: Телец\nСтихия: Земля\nПланета: Венера')
elif (month == 'май' and (21 <= day <= 31)) or (month == 'июнь' and (1 <= day <= 21)):
    print('Знак зодиака: Близнецы\nСтихия: Воздух\nПланета: Меркурий')    
elif (month == 'июнь' and (22 <= day <= 30)) or (month == 'июль' and (1 <= day <= 22)):
    print('Знак зодиака: Рак\nСтихия: Вода\nПланета: Луна') 
elif (month == 'июль' and (23 <= day <= 31)) or (month == 'август' and (1 <= day <= 22)):
    print('Знак зодиака: Лев\nСтихия: Огонь\nПланета: Солнце')
elif (month == 'август' and (23 <= day <= 31)) or (month == 'сентябрь' and (1 <= day <= 22)):
    print('Знак зодиака: Дева\nСтихия: Земля\nПланета: Меркурий')
elif (month == 'сентябрь' and (23 <= day <= 30)) or (month == 'октябрь' and (1 <= day <= 23)):
    print('Знак зодиака: Весы\nСтихия: Воздух\nПланета: Венера')
elif (month == 'октябрь' and (24 <= day <= 31)) or (month == 'ноябрь' and (1 <= day <= 22)):
    print('Знак зодиака: Скорпион\nСтихия: Вода\nПланета: Плутон')
elif (month == 'ноябрь' and (23 <= day <= 30)) or (month == 'декабрь' and (1 <= day <= 21)):
    print('Знак зодиака: Стрелец\nСтихия: Огонь\nПланета: Юпитер')
elif (month == 'декабрь' and (22 <= day <= 31)) or (month == 'январь' and (1 <= day <= 20)):
    print('Знак зодиака: Козерог\nСтихия: Земля\nПланета: Сатурн')
elif (month == 'январь' and (23 <= day <= 31)) or (month == 'февраль' and (1 <= day <= 18)):
    print('Знак зодиака: Водолей\nСтихия: Воздух\nПланета: Уран')
elif (month == 'февраль' and (23 <= day <= 28)) or (month == 'март' and (1 <= day <= 20)):
    print('Знак зодиака: Рыбы\nСтихия: Вода\nПланета: Нептун')
 
```
