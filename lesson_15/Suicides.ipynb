```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="darkgrid")

from pylab import rcParams
rcParams['figure.figsize'] = 10,7
```

```python
suicide = pd.read_csv('master.csv')
suicide.head()
```

## <center> График №1. Кто больше совершает самоубийств по гендерному признаку?

```python
sizes = suicide.groupby('sex')['suicides_no'].sum()
explode = (0.1, 0)
labels = ['female', 'man']
plt.pie(sizes, explode=explode, labels=labels,  autopct='%1.1f%%')
plt.show()
```

#### Больше всего самоубийств совершают мужчины, причем с очень большим большим отрывом, практически в 3 раза.

#### Использована круговая диаграмма, т.к количество отображаемых значений мало и они сильно отличаются друг от друга. В построении круговой диаграммы использована группировка и в качестве агрегации использована сумма.

## <center> График №2. Самоубийства в возрастных группах.

```python
x = suicide.groupby('age')['suicides_no'].sum().sort_values(ascending=False)
x.plot(kind='barh')
plt.xlabel('count')
plt.ylabel('age group')
plt.show()
```

#### Самое большое количество самоубийств в возрасной группе 35-54 года. Возможно это связано с "кризисом среднего возраста" или с нарастающими трудностями в семьи или работе.

#### Использована горизонтальная столбчатая диаграмма для сравнения числовых значений в разных возрастных группах. В построении диаграммы использована группировка и в качестве агрегации использована сумма.  

## <center> График №3 Самоубийства в разрезе ВВП на душу населения

```python
x = suicide.groupby('gdp_per_capita ($)')['suicides_no'].sum()
plt.hist(x, bins=100, range=(100,5000))
plt.ylabel('count')
plt.xlabel('gdp_per_capita ($)')
plt.grid()
plt.show()
```

## <center> График №4-5. В каких странах больше всего самоубийств за отчетный период?

```python
country = suicide[['country', 'suicides_no']]
country.head()
```

```python
countries = country.groupby('country').sum().sort_values(by='suicides_no', ascending=True)
countries = countries.loc[countries['suicides_no'] > 30000]
countries.plot(kind='barh')
```

__Больше всего самоубийств в РФ, также наблюдается относительно большое количество самоубийств в бывших республиках СССР(РФ, Украина, Казахстан, Беларусь, Узбекистан. Также много случаев в богатых странах типа США или Японии__

```python
count_suicide = suicide[['country', 'suicides/100k pop']]
result = count_suicide.groupby('country').agg({'suicides/100k pop': 'mean'})

result = result.loc[result['suicides/100k pop'] > 15].sort_values(by='suicides/100k pop', ascending=True)
result.plot(kind='barh')
```

__А вот по количеству самоубийств по 100к жителей, неожиданно лидирует Латвия и Шри-Ланка. Но все также в топ-10 вошли почти все крупные республики бывшего СССР, что в сумме с предыдущих графиком доказывает нам о очень большом количестве самоубийств в республиках бывшего СССР, как в абсотном так и относительном значении. Также в топе присутствует много азиатских стран(Южная Корея, Япония), несмотря на достаточно высокий уровень жизни, можно сказать, что в этих странах тоже много случаев суицида, связано это возможно с менталитетом людей. Он выражается в ответственности, "верности компании или работодателю", нам известны   частые случаи самоубийства на рабочем месте в этих странах__

## <center> График №6. Как менялся тренд с течением времени?

```python
years = suicide[['year', 'suicides_no']]
yearss = [year for year, df in suicide.groupby('year')]
plt.plot(yearss, years.groupby('year').sum())

plt.xlabel('years')
plt.ylabel('count')
plt.show()
```

#### Начиная с 1988 года число самоубиуств резко возросло. На следующей визуализации мы посмотрим в каких странах было больше всего самоубийств с 1988 по 1995 года, чтобы понять причины.

#### Здесь использован линейный график, т.к он наиболее хорошо отображает количественные показатели за непрерывный интервал времени. Для его создания использован List Comprehension и GroupBy.

## <center> График №7. Топ-5 стран по случаям самоубийства за 1988-1995гг.

```python
import squarify 
```

```python
top_5 = suicide[['country', 'suicides_no', 'year']]
top_5 = top_5.loc[top_5['year'].between(1988,1995)]
top_5 = top_5.groupby('country').sum().sort_values(by='suicides_no', ascending=False)
top_5 = top_5[['suicides_no']]
top_5 = top_5.head()

sizes = top_5['suicides_no']
label = ['1. Russia', '2. USA', '3. Japan', '4. Ukraine', '5. France']
color = ['#3c65be', '#438e56', '#c4a2eb', '#b48474', '#dcfa6f']

squarify.plot(sizes=sizes, label=label, alpha=.5, color=color)
plt.axis('off')
plt.show()
```

#### Итак, за период 1988-1995гг., когда был большой всплеск суицидов, большинство случаев приходиться на Россию, также отметим Украину в топе. Можно предположить, что это связано с развалом СССР, люди оказались в один миг в дгурой стране, которая была относилась к гражданам намного жестче, но в тоже время давала людям больше возможностей, отчасти (не бОльшая часть) люди которые не смогли себя найти, решали свести счеты с жизнью. Нахождение развитых стран в топе таких как США, Япония и Франция имеет разные причины. 

#### В данном случае использована древовидная диаграмма. Не лучший вариант в подобной ситуации, я бы остановился на столбчатой вертикальной диаграмме, но в условиях ДЗ надо использовать разные стили диаграмм.
